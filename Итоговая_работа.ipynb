```python
from sympy import *
```

```python
x= Symbol("x")
```

## Функция ##

```python
function = simplify(sin(x)/2 + cos(x)/2) ** 2 * sin(x)
function
```

## Определить корни

```python
kor = sorted(solveset(function, x, Interval(-pi, pi)))
pretty_print(kor)
```

## Определение положительных и отрицательных диапазонов значения функции

```python
for i in range(len(kor) - 1):
    if function.subs(x, ((kor[i+1] + kor[i])/2.0)) > 0:
        print(f'Функция положительна: {kor[i]} до {kor[i+1]}')
    else:
        print(f'Функция положительна: {kor[i]} до {kor[i+1]}')
```

## Определение производной

```python
dif_func = trigsimp(diff(function))
eq = Eq(dif_func, 0)
eq
```

```python
res_solve_dif_func = solve(dif_func, x)
for i in res_solve_dif_func:
    print(round(i.evalf(), 2))
```

## Построить график 

```python
plot(function, dif_func, (x, -pi, pi), title = "График функции и ее производной")
```

## Вычислить вершину

```python
for x in res_solve_dif_func:
    y = (sin(x)/2 + cos(x)/2) ** 2 * sin(x)
    print(f'x = {x}, y = {y}')
    print(f'x = {round(x.evalf(),2)}, y = {round(y.evalf(), 2)}')
    print()
```
